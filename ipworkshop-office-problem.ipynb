```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv('500_Person_Gender_Height_Weight_Index.csv')
df.head()
```

```python
# We don't know Gender, Weight, Index
df = df.drop(columns=['Gender', 'Weight', 'Index'])
df.describe()
```

```python
# Turn heights to an array

# .reshape() on x because this array is required to be two-dimensional
x = np.arange(1, (df.size + 1)).reshape((-1, 1))
y = df[['Height']].to_numpy()

# Apply linear regression to the model
model = LinearRegression()
model.fit(x,y)

r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)

print('intercept:', model.intercept_)

print('slope:', model.coef_)
```

```python
y_pred = model.predict(x) # predict =  model.intercept_ + model.coef_ * x
print('predicted response:', y_pred, sep='\n')
```

```python
# Example forecast for future heights
x_new = np.arange(5).reshape((-1, 1))
y_new = model.predict(x_new)
print(y_new)

```

```python
plt.scatter(x, y)

plt.axis([0, 500, 0, 280])
plt.xlabel('Number of people', fontsize=18)
plt.ylabel('Height of people', fontsize=16)

# draw intercept line
plt.plot([0, 500], [df["Height"].mean(), df["Height"].mean()], color='red')

plt.show()






```
